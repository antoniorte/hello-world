### Aguja de Buffon

El problema de la aguja de Buffon es en problema relacionado que habla sobre la teoria de la probabilidad que relaciona la geometría con el cálculo del número  π. Este fue formulado por el frances Goerge-Lous Leclerc, Conde de Buffon, durante el siglo XVIII. Donde la idea es lanzar una aguja con una longuitud sobre un plano marcado con líenas paralelas separadas por una distancia, donde el objetivo es calcular la probabilidad que la aguja cruce alguna de estás líenas.

**¿De que trata?**

Como se explico anteriormente, se trata de determinas el valor π lanzado una abuja sobre dos líneas, tratando de responder la pregunta: ¿Cuál es la probabilidad de que la aguja curce alguna de estas líenas?

Sin embargo, no solo engloba conceptos de probabilidad, sino, como anteriormente se dijo, también, engloba concetos de geometría y estádistica y además, esto se puede resolver mediente cálculo integral y con otro tipo de metodos. Por otro lado, ya que existe una relación entre la longitud de la aguja, la distancia entre las líenas y el número de cruces, hace que se puede obtener una formula que aproxima el valor de π partiendo de experimetos aleatorios.

**Solución con integrales**

Sobre el tema de las integrales, si se decidera resolver el problema de al aguja usando estó lo que estariamos haciendo es sacando el área bajo la curva, sin embargo, lo que determinan nuestros ejes es $l$ que es la longitud de la aguja y $d$ que es la separación entre las líenas. y la función que describe esto esto es:

$f(\theta) = \frac{l}{d} \sin(\theta)$

Donde el angulo es la inclinación de la aguja con respecto a la horizontal y al integral esto los valores varia entre 0 y π.

**Sin integrales**

Sin embargo, si no usaramos calculo integral para poder hacer esto tendría que ser con probabilidades, donde el cruce de las líenas es una relación entre las dos medidas y se tendrían que hacer $n$ lanzamientos y contar las veces que sí cruzó y no cruzó con el fin de poder sacar una estimación para π. Entonces esto se experesaría así:


$P= \frac{2l}{dπ} $

**Código**

```python
import numpy as np

def f1(theta, l, d):
  return (l / d) * np.sin(theta)
def monte1(f, l, d, n):
  theta=np.random.uniform(0, np.pi / 2, n)
  cruce=np.mean(f(theta, l, d)>=np.random.uniform(0, 1, n)) #compara valores
  return (2*l)/(d*cruce)
l=1.0
d=2
n=100000
monte1(f1, l, d, n)
```

- ¿Qué variables aleatorias hay en este caso?

  $\theta$ representa el angulo de inclinación que tiene la aguja con respecto de la líneas

  $l$ es la longitud de la cuerda
  
  $d$ es la longiurd de las lineas
