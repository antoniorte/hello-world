## Ejercicio 1

El perímetro de un triángulo isósceles es de 20 cm. Imagina que este triángulo rota centrado en su eje de simetría para formar la ilusión de un cono.

¿Cuánto deben medir sus lados para que el volumen generado por la rotación sea el mayor posible?

image.png

```python
# Notas:
# Vmax= pi/3r2h 
# Dibujo en el cuaderno
# Perimetro = 20=2y+2x=10=x+y 
# H= teromea de pitagoras x2+h2=y2 entonces = h=raiz(y2-x2)
# Radio este va hacer igual a x por lo que r2=x2
# h= Raiz(10-x)2-x2
# Entonces la función queda: pi/3*x2*raiz((10-x)2-x2)
# Sinmplificada queda como: (pi / 3) * x**2 *sqrt(100 - 20 * x)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.var("x")
sp.var("y")
sp.var("h")
y_respecto_a_x = sp.solve(2*x + 2*y - 20, y)[0]
altura = sp.solve(x**2 + h**2 - y**2, h)
altura_x = altura[1].subs(y, y_respecto_a_x)
altura_x
```

```python
Vmax=(sp.pi / 3) * x**2 * h
Vmax=Vmax.subs(h, altura_x).subs(y, y_respecto_a_x)
```

```python
dV=sp.diff(Vmax, x) #Derivar con respecto de x
pc=sp.solve(dV, x)[1]
```

```python
y_respecto_a_x.subs(x, pc)
```

El volumen máximo del cono se consigue cuando $x=8$ y $y=6$

## Ejercicio 2

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol en un parque. Es una portería simple que tendrá sólo dos postes y el travesaño:

Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?



```python
# Notas: 
# Una porteria es x*y
# Son tres palos entonces x+x+y=6
import numpy as np
import matplotlib.pyplot as plt
sp.var("x")
sp.var("y")

perimetro = 2*x + y - 6
A_xy = x*y
y_x = sp.solve(perimetro, y)[0]
A_x = A_xy.subs(y, y_x)
A_x
DA=sp.diff(A_x)
pc=sp.solve(DA, x)[0] #Encontre x
pc
y_x.subs(x, pc)
```

La medida máximo para los postes de la porteria son las siguientes: postes laterales $x=3/2$ y travesaño $y=3$

## Ejercio 3

**Resistencia de una viga**  Una viga de madera tiene una sección transversal rectangular de altura $h$ y ancho $w$. La resistencia $S$ de la viga es directamente proporcional al ancho y al cuadrado de su altura ($S=kwh^2$, donde $k$ es una constante de proporcionalidad). ¿Cuáles son las dimensiones de la viga más fuerte que puede cortarse a partir de un leño redondo de $24$ pulgadas de diámetro?

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

C = patches.Circle((0, 0),12, fill=False,color='brown',lw=4)
S = patches.Rectangle((-9,-8),18,16,fill=False,color='k')
ax = plt.gca()
ax.add_artist(S)
ax.add_artist(C)

plt.text(0,8.5,'$w$')
plt.text(9.5,0,'$h$')

plt.plot([-9,9],[-8,8],'r--')
plt.text(0,2,'$24$')

plt.axis('equal')
plt.ylim([-15,15])
plt.axis('off')
plt.show()
```

```python
import sympy as sp
sp.var("x")
sp.var("y")
Area=x*y
Diagonal=x**2+y**2-24**2
y_x=sp.solve(Diagonal, y)[1]
Area_x=Area.subs(y, y_x)
dA=sp.diff(Area_x)
pc=sp.solve(dA, x)[1]
y_x.subs(x, pc)
```

**Iluminación**  Una fuente luminosa se localiza sobre el centro de una mesa circular de $4 pies$ de diámetro. Encontrar la altura $h$ de la fuente luminosa de modo tal que la iluminación $I$ en el perímetro de la mesa sea máxima si $I=k(\sin \alpha)/s^2$, donde $s$ es la altura oblicua, $\alpha$ es el ángulo al cual la luz incide sobre la mesa y $k$ es una constante (El valor de esta última constante es irrelevante para el problema, ya que no se pide el valor de la iluminación).

```python
E = patches.Ellipse((0, 0),10, 3,fc='white',ec='brown')
C = patches.Circle((0,10.5),0.5,color='y')
a = patches.Arc((-5,0),2,2,theta2=60)
p1 = patches.Rectangle((-5, -2),0.5, 2, fill=False,color='brown')
p2 = patches.Rectangle((4.5, -2),0.5, 2, fill=False,color='brown')
p3 = patches.Rectangle((-0.25, -3),0.5, 2, fill=False,color='brown')
ax = plt.gca()
ax.add_artist(p1)
ax.add_artist(p2)
ax.add_artist(p3)
ax.add_artist(E)
ax.add_artist(C)
ax.add_artist(a)

plt.text(2,5,'$s$')
plt.text(0,3,'$h$')
plt.text(-4,0.5,r'$\alpha$')
plt.text(-0.5,-1,'$4$ pies')

plt.plot([-5,0,5],[0,10,0],'y--')
plt.plot([-5,5],[0,0],'k--')
plt.plot([0,0],[0,10],'k--')


plt.axis('equal')
plt.ylim([-3.1,12])
plt.axis('off')
plt.show()
```

```python
# Notas: 
# Para encontrar s puedo hacer teorema de pitagoras donde: s2=h2+2**2  entonces s = raiz(h2+4)
# Para encntrar el angulo puedo usar trigonometria donde seria 2/s por lo tanto seria 2/raiz(h2+4)
# Entonces la función de I quedaría como I=2(h2+4)3/2
#I = k sin(alpha)/s**2
#sin(alpha) = co/hipotenusa = h/s
#I = k sin(alpha)/s**2
sp.var("s")
sp.var("h")
S=-s**2 +h**2 + 2**2
s_h=sp.solve(S, s)[1]
Alpha=h/s
Alpha_s=Alpha.subs(s,s_h)
I=Alpha_s/s_h**2
dI=sp.diff(I, h)
pc=sp.solve(dI, h)[1] #Esto ya es h 
pc
```

La altura que deberia de tener la lampará es de $\sqrt2$
